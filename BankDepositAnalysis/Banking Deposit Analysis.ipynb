# Split the Data

### Split the Data to training set and test set, set the test size to 30%
To make sure there’s no issues between Pandas data type and the Numpy numeric, we can request the values.<br>
Add .values -> for example: y = df['price'].values<br>
If you state values, it returns back the Numpy array underneath the actual data frame.

```python
X = df.drop('y', axis=1) 

```

```python
y = df['y'] 
```

### Convert categorical columns to dummy variables


```python
X = pd.get_dummies(X, drop_first=True)
```

# Algorithm

### Choose an algorithm to check for the prices prediction, fit and predict
Decide what are the appropriate algorithm to solve this kind of problem<br> - use logistic regration and liniar regretion

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=207408022)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## <u>Logistic Regression</u>

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# הגדרת פרמטרים לחיפוש
param_grid = {
    'C': [0.1, 0.5, 0.75, 1, 2],  # ערכי רגולריזציה
    'solver': ['newton-cg', 'sag', 'saga', 'lbfgs'],  # אלגוריתמים לפתרון
    'max_iter': [50, 100, 200],  # מספר האיטרציות
}

# יצירת המודל
logreg = LogisticRegression(random_state=207408022)

# הגדרת GridSearchCV
grid_logreg = GridSearchCV(estimator=logreg, param_grid= param_grid, verbose=3, scoring='accuracy', cv=5)

#אימון המודל
grid_logreg.fit(X_train, y_train)


```

```python
grid_logreg.best_estimator_
```

```python
logreg = LogisticRegression(C=2, max_iter=50, solver='sag', random_state=207408022)

logreg.fit(X_train, y_train)
```

### Predict class labels for samples in x_test 

```python
predictions_logreg = logreg.predict(X_test)
```

### Print the accuracy for the training set

```python
print('Train set score: {:.4f}'.format(logreg.score(X_train, y_train)*100))
```

### Print the accuracy for the test set

```python
print('Test set score: {:.2f}'.format(logreg.score(X_test, y_test)*100))
```
